# 3. 작품의 수가 많은 순서대로 나라를 출력하고 해당 나라에서 출시한 작품의 수를 나라의 이름과 함께 각각 출력

```python
arr_country = df['country']  # country확인
arr_country 
```

```python
# country를 ','를 기준으로 분리해주기
arr_country = df['country'].str.split(', ', expand=True).stack().reset_index(level=1,)
arr_country
```

### 모든 나라 및 작품 개수 출력

#### 리스트로 만들어주기

```python
list_country = list(np.array(arr_country[0].tolist()))  
list_country
```

#### 각 나라의 작품 개수

```python
country_count={}
for i in list_country:
    try: country_count[i] += 1
    except: country_count[i]=1
country_count
```

#### 작품 개수가 가장 많은 나라부터 출력 (정렬)

```python
 dict(sorted(country_count.items(), key=lambda item: item[1], reverse=True))
```

# 4. 넷플릭스 작품들에 많이 나타나는 키워드 추출

#### description를 ' '를 기준으로 분리해주기

```python
arr_description = df['description'].str.split(' ',expand=True).stack()
arr_description
```

#### 영어를 제외한 문자 삭제 (. , " "  ? ! _)

```python
arr_description = arr_description.str.rstrip('.')  # 마침표 지워주기
arr_description = arr_description.str.rstrip(',')  # 쉼표 지워주기
arr_description = arr_description.str.strip('_')  # 물음표 뒤에 '_'지워주기
arr_description = arr_description.str.strip('"')  # 큰 따옴표 지워주기
arr_description = arr_description.str.strip('?')  # 물음표 지워주기
arr_description = arr_description.str.rstrip('!')  # 물음표 지워주기
arr_description
```

```python
arr_description = arr_description.str.split(' ',expand=True).stack().reset_index(level=1,)
arr_description
```

#### 리스트로 바꿔주기

```python
list_description = list(np.array(arr_description[0].tolist()))  
list_description 
```

#### 대문자는 모두 소문자로 바꿔주기

```python
c_lower= []
for i in list_description:
    c_lower.append(i.lower())
    
c_lower
```

#### 전치사, to, for, as, the, a, at, of, on, in, into, with, and, but, however 등 제거

```python
remove_set = {'is', 'are', 'to', 'for', 'as', 'the', 'a','an', 'at', 'of', 'on', 'in', 'into', 'with', 'and', 'but', 'however'}
description= [i for i in c_lower if i not in remove_set]
description
```

#### 같은 단어들끼리 모아주기

```python
description_count={}
for i in description:
    try: description_count[i] += 1
    except: description_count[i]=1
description_count
```

#### 가장 많이 나타나는 키워드

```python
max(description_count, key = description_count.get)
```
